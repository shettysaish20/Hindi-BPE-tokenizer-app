## Building Regex string

```python
import re

hindi_text = "भारत की राजधानी दिल्ली है। २०२५ में नई योजनाएँ लागू होंगी।                           AI और मशीन लर्निंग महत्वपूर्ण विषय हैं! वर्ष 2025 भी महत्वपूर्ण होगा।"

# Updated regex pattern
# pattern = r"[\u0900-\u097F]+|[\u0966-\u096F]{1,3}|[\d]{1,3}|[a-zA-Z]+|[\u0964\u0965]|[^\s\w\u0900-\u097F]"
# pattern = r"\u0964"
# pattern = r"[\u0900-\u097F]+(?=[\s\u0964\u0965])|[\u0966-\u096F]{1,3}|[\d]{1,3}|[a-zA-Z]+|[\u0964\u0965]|[^\s\w\u0900-\u097F]"
## Updated regex pattern combining hindi characters and gpt tokenizer details
# pattern = r"\s?[\u0900-\u0963\u097A-\u097F]+|[\u0966-\u096F]{1,3}|[\d]{1,3}|\s?[a-zA-Z]+|[\।.,!?;:\"'(){}[\]-\u0964\u0965]|[^\s\w\u0900-\u097F]|\s*[\r\n]| \s+(?!\S)|\s+"

# Final Regex pattern (combining hindi characters and draws inspiration from GPT-4 tokenizer)
pattern = r"(?i:[sdmt]|ll|ve|re)|\s?[\u0900-\u0963\u097A-\u097F]+|[\u0966-\u096F]{1,3}|[\d]{1,3}|\s?[a-zA-Z]+|[।.,!?;:\"'(){}[\]-\u0964\u0965]|[^\s\w\u0900-\u097F]|\s*[\r\n]|\s+(?!\S)|\s+"

# Splitting text
tokens = re.findall(pattern, hindi_text)

print(tokens)
```
